# Задание 1. Словари и множества
Написать программу, которая подсчитывает количество вхождений каждого символа в заданной строке и выводит результат в виде словаря.

```python
def count_chars(text):
    char_counts = {}
    for char in text:
        if char not in char_counts:
            char_counts[char] = 1
        else:
            char_counts[char] += 1
    return char_counts

text = input("Введите строку: ")
char_counts = count_chars(text)
print(f"Количество символов в строке: {char_counts}")
```

# Задание 2: Словари и множества
Написать программу, которая создает множество из списка чисел, введенных пользователем и выводит его сначала в неизмененорм виде, а затем удаляет дубликаты и выводит результат без дублей.

```python
def print_array_set(numbers):
    array = numbers.split()
    print("Исходное множество:", array)
    set_numbers = set(map(int, array))
    print("Множество без дубликатов:", set_numbers)


numbers = str(input("Введите числа через пробел: "))
print_array_set(numbers)
```

# Задание 3: Модули и пакеты
Написать программу для отображения текущей даты и времени в формате "15 февраля 2025, 12:34:56"

```python
import datetime

def now_date_time():
    current_datetime = datetime.datetime.now()
    formatted_date = current_datetime.strftime("%d %B %Y, %H:%M:%S")
    print(formatted_date)

now_date_time()
```

# Задание 4: Модули и пакеты
1. Создайте программу, которая просит пользователя ввести имя файла изображения.
2. Загрузите изображение
3. Измените размер изображения, уменьшив его вдвое.
4. Сохраните изменнное изображение под новым именем.
5. Выведите информацию о размере исходного и изменного изображений.

```python
from PIL import Image

def new_size_image(filename, new_filename):
    img = Image.open(filename)
    display(img)
    width, height = img.size
    new_size = (int(width / 2), int(height / 2))
    new_image = img.resize(new_size)
    new_image.save(new_filename)
    display(new_image)

filename = input("Введите имя файла")
new_filename = input("Введите новое имя файла")
new_size_image(filename, new_filename)
```
